```python
import pandas as pd
import os
# pd.options.mode.chained_assignment = None

os.makedirs("results/inputs_csv", exist_ok=True)

csv_files = sorted(os.listdir("results/inputs_csv"))
print(csv_files)

for i in range(len(csv_files)):
    csv_file_name = csv_files[i]
    print(csv_file_name)
    df = pd.read_csv(f"results/inputs_csv/{csv_file_name}")

    df_prices = df[["maj", "nom", "valeur"]]

    # type carburant par dates
    print(df_prices["nom"].unique())

    list_carburant = df_prices["nom"].unique()

    for carburant in list_carburant:
        # if carburant == None:
        #     break
        df_carburant = df_prices[df_prices["nom"]== carburant]
        # print(df_carburant[:5])
    
        df_carburant = df_carburant.sort_values(by=['valeur'])
        print("valeur max")
        print(df_carburant.nlargest(5, 'valeur'))

        print("valeur min")
        print(df_carburant.nsmallest(10, 'valeur'))
```
