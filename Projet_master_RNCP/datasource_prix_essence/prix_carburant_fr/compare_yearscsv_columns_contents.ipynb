##### On veut comparer les csv de chaque années sur certaines colonnes voir si il s'agit des mêmes d'années en années.

```python
import pandas as pd
pd.options.mode.chained_assignment = None
import os
```

```python
def verify_similarity(df_databrick, df_menaps, columns):
    for column in columns:
        df_databrick[column] = df_databrick[column].astype(str)
        df_menaps[column] = df_menaps[column].astype(str)
    
    df_merge = pd.merge(df_databrick, df_menaps[columns], on=columns, how='left', indicator=True)
    df_only_databrick = df_merge[df_merge['_merge'] == 'left_only']
    df_only_databrick['_merge'] = 'PRECEDENT_ONLY'
    print("values ONLY PRECEDENT = ", len(df_only_databrick))

    df_merge2 = pd.merge(df_menaps, df_databrick[columns], on=columns, how='left', indicator=True)
    df_only_menaps = df_merge2[df_merge2['_merge'] == 'left_only']
    df_only_menaps['_merge'] = 'ACTUAL_ONLY'
    print("values ONLY ACTUAL = ",len(df_only_menaps))

    df_both = df_merge[df_merge['_merge'] == 'both']
    print("values BOTH into first merge = ", len(df_both))
    df_both2 = df_merge2[df_merge2['_merge'] == 'both']
    print("values BOTH into second merge = ", len(df_both2))
    df_both['_merge'] = 'BOTH'
    
    result_concat = pd.concat([df_only_databrick, df_only_menaps, df_both], ignore_index=True)
    
    return result_concat
```

```python
os.makedirs("results/inputs_csv", exist_ok=True)

csv_files = sorted(os.listdir("results/inputs_csv"))
print(csv_files)

for i in range(len(csv_files) - 1):
    precedent_year = csv_files[i]
    # precedent_year = "PrixCarburants_annuel_2009.csv"
    actual_year = csv_files[i+1]
    print(precedent_year, actual_year)

    precedent_df = pd.read_csv(f"results/inputs_csv/{precedent_year}")
    actual_df = pd.read_csv(f"results/inputs_csv/{actual_year}")

    
    precedent_unique = precedent_df[["id", "cp", "ville", "adresse"]].drop_duplicates()
    print("precedent_unique", len(precedent_df[["id", "cp", "ville", "adresse"]].drop_duplicates()))

    actual_unique = actual_df[["id", "cp", "ville", "adresse"]].drop_duplicates()
    print("actual_unique", len(actual_df[["id", "cp", "ville", "adresse"]].drop_duplicates()))

    test_similarity = verify_similarity(precedent_unique, actual_unique, ["id", "cp", "ville", "adresse"])
    
    os.makedirs("results/results_comparaison", exist_ok=True)
    test_similarity.to_csv(f"results/results_comparaison/compare_{csv_files[i]}{csv_files[i+1]}.csv", index=False)
```
